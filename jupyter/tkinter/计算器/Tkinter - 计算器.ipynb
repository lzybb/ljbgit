```python
# 添加逻辑功能

from tkinter import *

root = Tk()

# 定义面板的大小
root.geometry('250x380')
root.title('ljb')
# 定义面板
# bg背景颜色
# #dddddd是十六进制表示颜色的一个串
frame_show = Frame(width=300,height=200,bg='#dddddd')
frame_show.pack()

# 定义顶部区域

sv = StringVar()
sv.set('0')

# anchor:定义控件的瞄点
# font代表字体
show_label = Label(frame_show, textvariable=sv,
                  bg='green',width=12,height=1,
                  font=('黑体',20, 'bold'),
                  justify=LEFT,anchor='e')

show_label.pack(padx=10,pady=10)

def delete():
    pass


# 按键区域
frame_bord = Frame(width=400,height=350,bg='#cccccc')

b_del = Button(frame_bord,text='←',width=5,height=1,
              command=delete)
b_del.grid(row=0,column=0)

同理 清除，关闭,±

'''
考虑几种情况：
1.按下数字
2.按下操作符号
3.只考虑两个操作数操作

'''
num1 = ''
num2 = ''
operator = ''



def change(num):
    '''
    按下一个数字需要考虑两种情况：
    1.数字属于第一个操作数
    2.数字属于第二个操作数
    3.判断是否属于第一个操作数，可以通过operator判断   
    '''
    # 假如操作数是None,表明肯定是第一个操作数
    if operator == None: # if not operator
        num1 += num
        # 如果是第一个操作数，则只显示第一个操作数
        sv.set(num1)
    else:
        num2 += num2 
        # 如果是第二个操作数，则应该是显示完整的计算式子
        sv.set(num1+operator+num2)
        
def operation(op):
    if op in ['+','-','*','/']:
        operator = op
    else: # 认为是按下的是等号
        if op == '+':
            rst = int(num1) + int(num2)
        if op == '-':
            rst = int(num1) - int(num2)
        if op == '*':
            rst = int(num1) * int(num2)
        if op == '/':
            rst = int(num1) / int(num2)
            
        sv.set(str(rst))

b_1 = Button(frame_bord,text='1',width=5,height=2,command=lambda:change('1'),)

b_1.grid(row=1,column=0)

其他数字同理

运算符号



b_jia = Button(frame_bord,text='+',width=5,height=2,command=lambda:operation('+'),)

b_jia.grid(row=1,column=4)

其他同理



frame_bord.pack(padx=10, pady=10)

root.mainloop()
```
