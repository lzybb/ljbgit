### 组件的大致使用步骤
1. 创建总面板
2. 创建面板上的各种组件
    1. 指定组件的父组件，即依附关系
    2. 利用相应的属性对组件进行设置
    3. 给组件安排布局

3. 同步骤2相似，创建好多个组件
4. 最后，启动总面板的消息循环

### 组件布局
- 控制组件的摆放方式
- 三种布局：
    - pack:按照方位布局
    - place: 按照坐标布局
    - grid: 网格布局
    
- pack布局
    - 最简单，代码量最少，挨个摆放，默认从上到下，系统自动设置
    - 通用使用方式为：组件对象.pack(设置.....)
    - side:停靠方位，可选值为：LEFT,TOP,RIGHT,BOTTOM
    - fill：填充方式，X,Y,BOTH,NONE
    - expande:YES/NO
    - anchor：N，E,S,W,CENTER
    - ipadx: x方向的内边距
    - ipady：y方向的内边距
    - padx： x方向外边界
    - pady： y方向外边界 

- grid布局
    - 通用使用方式：组件对象.grid(设置.....)
    - 利用row,column编号，都是从0开始
    - sticky: N,E,S,W表示上(N)下(S)左(W)右(E)，用来决定从哪个方向开始
    - 支持ipadx,padx等参数，跟pack函数含义一样
    - 支持rowspan,columnspan,表示跨行，跨列数量

- place布局
    - 明确方位的摆放
    - 相对位置布局，随意改变窗口大小会导致混乱
    - 使用place函数，分为绝对布局和相对布局，绝对布局使用x,y参数
    - 相对布局使用relx,rely,relheight,relwidth

```python
# pack布局案例
import tkinter
bf = tkinter.Tk()

# 以下所有代码都是创建一个组件，然后布局
b1 = tkinter.Button(bf,text='A')
b1.pack(side=tkinter.LEFT,expand=tkinter.YES,fill=tkinter.Y)

b2 = tkinter.Button(bf,text='B')
b2.pack(side=tkinter.TOP,expand=tkinter.YES,fill=tkinter.BOTH)

b3 = tkinter.Button(bf,text='C')
b3.pack(side=tkinter.RIGHT,expand=tkinter.YES,fill=tkinter.NONE,anchor=tkinter.NE)

b4 = tkinter.Button(bf,text='D')
b4.pack(side=tkinter.LEFT,expand=tkinter.NO,fill=tkinter.Y)

b5 = tkinter.Button(bf,text='E')
b5.pack(side=tkinter.TOP,expand=tkinter.NO,fill=tkinter.BOTH)

b6 = tkinter.Button(bf,text='F')
b6.pack(side=tkinter.BOTTOM,expand=tkinter.NO,fill=tkinter.X)

b7 = tkinter.Button(bf,text='G')
b7.pack(anchor=tkinter.SE) # side 默认参数是TOP

bf.mainloop()
```

## 消息机制
- 消息的传递机制
    - 自动发出事件/消息
    - 消息由系统负责发送到队列
    - 由相关组件进行绑定/设置
    - 后端自动选择感兴趣的事件并做出相应反应
- 消息格式：
    - <[modifier-]---type-[-detail]>
    - <Button-1>:Button表示一个按钮事件，1代表的是鼠标左键，2代表中键
    - <KeyPress-A>:键盘A键位
    - <Control-Shift-KeyPress-A>:同时按下Control,Shift,A三个键位
    - <F1>:F1键盘
    - [键位对应名称](https://infohost.nmt.edu/tcc/help/pubs/tkinter/web/key-names.html)
    - 中文的资料：
        -  https://blog.csdn.net/qin_shang_/article/details/79330466
        
    


```python
# 事件的简单例子
import tkinter

def bl(event):
    global bf
    lb = tkinter.Label(bf,text='谢谢点击')
    lb.pack()
# 程序的总框架    
bf = tkinter.Tk()

lb = tkinter.Label(bf,text='模拟按钮')
# label绑定相应的消息和处理函数
# 自动获取左键点击，并启动相应的处理函数bl
lb.bind('<Button-1>',bl)
lb.pack()

# 启动消息循环
# 到此，表示程序开始运行
bf.mainloop()
```

### Tkinter的绑定
- bind_all: 全局范围的绑定,默认的示全局快捷键，比如F1是帮助文件
- bind_class:接受三个参数，第一是类名，第二个是事件，第三个是操作
    - w.bind_class('Entry','<Control-V>',my_paste)
- bind:单独对某一个实例绑定
- unbind：解除绑定，需要一个参数，即你要解除哪个事件
    
### Entry
- 输入框，功能单一
- entry['show']='*',设置遮挡字符 

```python
# 输入框案例
import tkinter

def reg():
    # 从输入框中，得到相应输入
    name = e1.get()
    pwd = e2.get()
    
    t1 = len(name)
    t2 = len(pwd)
    
    if name == '111' and pwd =='222':
        lb3['text'] = 'sb！！！提示你也信'
        e1.delete(0,t1)
        e2.delete(0,t2)
    elif name == 'ljb' and pwd == '37':  # 此处用elif （if出现不判断的情况)
        lb3['text'] = '登入成功'
    else:
        lb3['text'] = '不对！请看提示:账户111,密码222'
        e1.delete(0,t1)  # 表示的是从0-t1之间的删除
        e2.delete(0,t2)
        
bf = tkinter.Tk()

lb1 = tkinter.Label(bf,text='用户名：',font='16').grid(row=0,column=0,sticky=tkinter.E)
e1 = tkinter.Entry(bf)
e1.grid(row=0,column=1,sticky=tkinter.W)

lb2 = tkinter.Label(bf,text='密码：',font='16').grid(row=1,column=0,sticky=tkinter.E)
e2 = tkinter.Entry(bf)
e2.grid(row=1,column=1,sticky=tkinter.W)
e2['show'] = '*'

bt = tkinter.Button(bf,text='登入',command=reg,font='16').grid(row=2,column=1) 
# sticky 没有center

lb3 = tkinter.Label(bf,text='',font='16')
lb3.grid(row=3,column=1)

bf.mainloop()
```

### 菜单
### 1.普通菜单
- 第一个Menu类定义的是parent
- add_command 添加菜单项，如果菜单是顶层菜单，则从左向右添加，否则就是下拉菜单
    - label: 指定菜单项名称
    - command:点击后相应的调用函数
    - acceletor：快捷键
    - underline：制定是否菜单信息下有横线
    - menu：属性制定使用哪一个作为顶级菜单

```python
# 普通菜单的代码
import tkinter
bf = tkinter.Tk()
menubar = tkinter.Menu(bf)
for i in ['File','Edit','View','About']:
    menubar.add_command(label=i)
bf['menu'] = menubar

bf.mainloop()
```

###  级联菜单
- add_cascade：级联菜单，作用是引出后面的菜单
- add_cascade的menu属性：指明把菜单级联到哪个菜单上
- label：名称
- 过程：
    1.建立menu实例
    2.add_command
    3.add_cascade

```python
import tkinter
# bf 是baseFrame的缩写

bf = tkinter.Tk()
menubar = tkinter.Menu(bf)
emenu = tkinter.Menu(menubar)

for i in ['Copy','Past','Cut']:
    emenu.add_command(label=i)
menubar.add_cascade(label='File')
menubar.add_cascade(label='Edit',menu=emenu)
menubar.add_cascade(label='View')

bf['menu'] = menubar

bf.mainloop()
```

```python
# 自我的失败案例：原意是将子级菜单添加到已有的父级菜单中的某个标签中
import tkinter
bf = tkinter.Tk()
menubar = tkinter.Menu(bf)
for i in ['File','Edit','View']:
    menubar.add_command(label=i)
emenu = tkinter.Menu(menubar)
for i in ['Copy','Past','Cut']:
    emenu.add_command(label=i)
menubar.add_cascade(label='Edit',menu=emenu)
bf['menu'] = menubar
bf.mainloop()
```

### 弹出式菜单
- 弹出式菜单也叫上下文菜单
- 实现的大致思路
    1.建立菜单并向菜单添加各种功能
    2.监听鼠标右键
    3.如果右键点击，则根据位置判断弹出
    4.调用Menu的pop方法
- add_separator: 添加分隔符

```python
import tkinter

def makeLabel():
    global bf
    tkinter.Label(bf,text='学历python中').pack()

bf = tkinter.Tk()
menubar = tkinter.Menu(bf)
for i in ['麻辣香菇','地锅鸡','青菜豆腐']:
    menubar.add_separator()
    menubar.add_command(label=i)
menubar.add_command(label='重庆火锅',command=makeLabel)

#事件处理函数一定要至少有一个参数，且第一个参数表示的是系统事件
def pop(event):
    # 注意使用 event.x 和 event.x_root的区别
    menubar.post(event.x_root, event.y_root)

bf.bind('<Button-3>',pop)

bf.mainloop()
```

### canvas 画布
- 画布：可以自由的在上面绘制图形的一个小舞台
- 在画布上绘制对象，通常create_xxx，xxx=对象类型， 例如line,rectangle
- 画布的作用是把一定组件画到画布上显示出来
- 画布所支持的组件：
    - arc （圆或者弧）
    - bitmap (位图)
    - image (BitmapImage, PhotoImage)
    - line（线）
    - oval(椭圆)
    - polygon(多边形)
    - rectangle(长方形)
    - text
    - windows(组件)
- 每次调用create_xxx都会返回一个创建的组件的ID，同时也可以用tag属性指定其标签
- 通过调用canvas.move实现一个一次性动作

```python
# 简单画布
import tkinter
bf = tkinter.Tk()

cvs = tkinter.Canvas(bf,width=300,height=200)
cvs.pack()

# 参数数字的单位是px(象素)
cvs.create_line(23,23, 178,245)# 线的端点的坐标
cvs.create_text(56,67, text='I am ljb ')# 文字书写起始的坐标点

bf.mainloop()

```

```python
# 画一个五角星
import tkinter
import math as m

bf = tkinter.Tk()

w = tkinter.Canvas(bf,width=300,height=300,background='gray')
w.pack()
center_x=150
center_y=150

r = 150

points=[
    # 左上点
    # pi就是π
    center_x - int(r*m.sin(2*m.pi/5)),
    center_y - int(r*m.cos(2*m.pi/5)),
    
    # 右上点
    center_x + int(r*m.sin(2*m.pi/5)),
    center_x - int(r*m.cos(2*m.pi/5)),
    
    # 左下点
    center_x - int(r*m.sin(m.pi/5)),
    center_y + int(r*m.cos(m.pi/5)),
        
    # 顶点
    center_x,
    center_y - r,
    
    # 右下点
    center_x + int(r*m.sin(m.pi/5)),
    center_y + int(r*m.cos(m.pi/5))
]

# 创建一个多边形
w.create_polygon(points,outline='green',fill='yellow')
w.create_text(150,150,text='五角星')

bf.mainloop()
```

```python
import tkinter
bf = tkinter.Tk()
def btnClick(event):
    global w
    w.move(id_ball,12,5)
    w.move('fall',0,5)
    
def btnClick1(event):
    global w
    w.move(id_ball,-12,-5)
    w.move('fall',0,-5)
    
w = tkinter.Canvas(bf,width=500,height=400)
w.pack()
w.bind('<Button-1>',btnClick)
w.bind('<Button-3>',btnClick1)

# 创建组件后返回id
id_ball = w.create_oval(3,123, 73,3 ,fill='green') # 椭圆相切形成的长方形的两个对角线点

# 创建组件使用tag属性
w.create_text(123,56,fill='red',text='LJB',tag='fall')
# 创建的时候如果没有指定tag可以利用addtag_withtag添加
# 同类函数还有 addtag_all, addtag_above, addtag_xxx等等
id_rectangle = w.create_rectangle(56,78,173,110,fill='gray')
w.addtag_withtag('fall',id_rectangle)

# tag 标签可以有好几个组件


bf.mainloop()
```
