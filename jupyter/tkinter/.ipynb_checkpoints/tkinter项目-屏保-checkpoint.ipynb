# Tkinter项目实战-屏保
### 项目分析
- 屏保可以自己启动，也可以手动启动
- 一旦敲击键盘或者移动鼠标后，或者其他的引发时间，则停止
- 如果屏保是一副画的话，则没有画框
- 图像的动作是随机的，具有随机性，可能包括颜色，大小，多少，运动方向，变形等
- 整个世界的构成是：
    - ScreenSaver: (屏保)
        - 需要一个canvas(画板),大小跟屏幕一致，没有边框
        
    - Ball
        - 颜色，大小，多少，运动方向，变形等随机
        - 球能动，可以被调用

```python
import random
import tkinter

class RandomBall():
    '''
    定义运动的球的类
    '''
    def __init__(self, canvas, scrnwidth, scrnheight):
        
        '''
        canvas：画布，所有内容都应该在画布上呈现出来，此处通过此变量传入
        scrnwidth/scrnheight:屏幕宽高
        '''
        # 球出现的初始位置要随机
        self.xpos = random.randint(15, int(scrnwidth)-20)
        self.ypos = random.randint(15, int(scrnwidth)-20)
    
        # 定义球运动的速度
        # 模拟运动：不断的擦掉原来的画，然后在一个新的地方再从新画
        # 此处xvelocity模拟x轴方向运动
        self.xvelocity = random.randint(4,20)
        self.yvelocity = random.randint(4,20)
    
        # 定义屏幕的大小
        self.scrnwidth = scrnwidth
        self.scrnheight = scrnheight
    
        # 球的大小随机
        # 此处球的大小用半径表示
        self.radius = random.randint(20,120)
    
        # 定义颜色
        # RGB表示法：三个数字，每个数字的值是0-255之间，表示红绿蓝的大小
        # 在某些系统中，之间用英文单词表示也可以，比如red,green,,blue
        # 此处用lambda表达式
        c = lambda:random.randint(0,255)
        self.color = '#%02x%02x%02x'%(c(),c(),c()) # 将c()的值转成16进制，并且显示两位，不足两位前面补0
        
    def create_ball(self):
        '''
        用构造函数定义的变量，在canvas上画一个球
        '''
        # tkinter没有画圆形函数
        # 只有一个画椭圆函数，画椭圆需要定义两个坐标
        # 在一个长方形内画椭圆，我们只需要定义长方形左上角和右下角就好
        x1 = self.xpos - self.radius
        y1 = self.ypos - self.radius
        
        x2 = self.xpos + self.radius
        y2 = self.ypos + self.radius
        
        # 在有两个对角坐标的前提下，可以画圆了
        # fill表示填充颜色
        # outline是外围边框颜色
        sefl.item = self.canvas.create_oval(x1,y1,x2,y2,
                                            fill=self.color,
                                           outline=self.color)
        
    def move_ball(self):
                
        # 移动球的时候，需要控制球的方向
        # 每次移动后，球都有一个新的坐标
        self.xpos += self.xvelocity
        self.ypos += self.yvelocity
        
        # 以下判断是否会撞墙
        if (self.xpos+self.radius)>self.scrnwidth or (self.xpos-self.radius)<0:
            self.xvelocity = self.xvelocity*(-1)
            
        if (self.ypos+self.radius)>self.scrnheight or (self.ypos-self.radius)<0:
            self.yvelocity = self.yvelocity*(-1)  
    
        # 在画布上挪动图画
        self.canvas.move(self.item, self.xvelocity, self.yvelocity)

    
    
class ScreenSaver():
    '''
    定义屏保的类
    可以被启动
    '''
    # 如何装随机产生的球？
    balls = list()
    
    def __init__(self):
        # 每次启动球的数量随机
        self.num_balls = random.randint(6,18)
        
        self.root = tkinter.Tk()
        # 取消边框
        self.root.overrideredirect(1)
        
        # 任何鼠标移动都需要取消
        self.root.bind('<Motion>',self.ljbquit)
        # 同理，按下任何键盘都退出屏保
        self.root.bind('<Key>',self.ljbquit)
        
        # 得到屏幕大小规格
        w,h = self.root.winfo_screenwidth(),self.root.winfo_screenheight()
        
        # 创建画布，包括画布的归属，规格
        self.canvas = tkinter.Canvas(self.root, width=w, height=h)
        self.canvas.pack()
        
        # 在画布上画球
        for i in range(self.num_balls):
            ball = RandomBall(self.canvas, scrnwidth=w, scrnheight=h)
            ball.create_ball()
            self.balls.append(ball)
        
        self.run_screen_saver()
        self.root.mainloop()
        
    def run_screen_saver():
        for ball in self.balls:
            ball.move_ball()
        # after是200毫秒后启动一个函数，需要启动的函数是第二个参数
        self.canvas.after(200,self.run_screen_saver)
        
    def ljbquit(self,event):
        # 此处只是利用了事件处理机制
        # 实际上并补关心事件的类型
        # 作业：
        # 此屏保程序扩展成，一旦捕获事件，则判断屏保不退出
        # 显示一个Button,Button上显示事件类型，点击Button后屏保才退出
        self.root.destroy()
       
    


if __name__== '__main__':
    #启动屏保
    ScreenSaver()
```

```python
# 成功案例-1

import random
import tkinter

class RandomBall():

    def __init__(self, canvas, scrnwidth, scrnheight):        
        self.canvas = canvas  # 此处赋值要有，原因为：第二个参数canvas = self.canvas,需要赋值等同回来。
        # 如果没有此赋值，后续的self.canvas需要改为canvas
        
        self.xpos = random.randint(15, int(scrnwidth)-20)
        self.ypos = random.randint(15, int(scrnwidth)-20)
    
        self.xvelocity = random.randint(4,20)
        self.yvelocity = random.randint(4,20)
    
        self.scrnwidth = scrnwidth
        self.scrnheight = scrnheight
    
        self.radius = random.randint(20,120)
    
        c = lambda:random.randint(0,255)
        self.color = '#%02x%02x%02x'%(c(),c(),c()) 
        
    def create_ball(self):
        
        x1 = self.xpos - self.radius
        y1 = self.ypos - self.radius
        
        x2 = self.xpos + self.radius
        y2 = self.ypos + self.radius
        
        self.item = self.canvas.create_oval(x1,y1,x2,y2,fill=self.color,outline=self.color)
        
    def move_ball(self):
                
        self.xpos += self.xvelocity
        self.ypos += self.yvelocity
        
        if (self.xpos+self.radius)>self.scrnwidth or (self.xpos-self.radius)<0:
            self.xvelocity = self.xvelocity*(-1)
            
        if (self.ypos+self.radius)>self.scrnheight or (self.ypos-self.radius)<0:
            self.yvelocity = self.yvelocity*(-1)  
    
        self.canvas.move(self.item, self.xvelocity, self.yvelocity)

    
    
class ScreenSaver():
    balls = list()   
    def __init__(self):
        self.num_balls = random.randint(6,18)        
        self.root = tkinter.Tk()
        self.root.overrideredirect(1)
        self.root.bind('<Motion>',self.ljbquit)
        self.root.bind('<Key>',self.ljbquit)       
        w = self.root.winfo_screenwidth()
        h = self.root.winfo_screenheight() 
        self.canvas = tkinter.Canvas(self.root, width=w, height=h)
        self.canvas.pack()

        for i in range(self.num_balls):
            ball = RandomBall(self.canvas, scrnwidth=w, scrnheight=h)
            ball.create_ball()
            self.balls.append(ball)
        
        self.run_screen_saver()
        self.root.mainloop()
        
    def run_screen_saver(self):
        for ball in self.balls:
            ball.move_ball()
        self.canvas.after(200,self.run_screen_saver)
        
    def ljbquit(self,event):
        self.root.destroy()

if __name__== '__main__':
    ScreenSaver()
    
```

```python
# 作业：
# 此屏保程序扩展成，一旦捕获事件，则判断屏保不退出
# 显示一个Button,Button上显示事件类型，点击Button后屏保才退出
import tkinter
import random

class rball(): 
    def __init__(self, canvas, canvas_w,canvas_h ):
        self.canvas = canvas  
        self.canvas_w = canvas_w
        self.canvas_h = canvas_h
        self.radius = random.randint(20,120)
        self.xpos = random.randint(self.radius,(self.canvas_w-self.radius)) 
        self.ypos = random.randint(self.radius,(self.canvas_h-self.radius))
        self.xmove = random.randint(5,25)
        self.ymove = random.randint(5,25)        
        x1 = self.xpos - self.radius
        y1 = self.ypos - self.radius      
        x2 = self.xpos + self.radius
        y2 = self.ypos + self.radius
        c = lambda: random.randint(0,255)
        self.color = '#%02x%02x%02x'%(c(),c(),c())
        self.color1 = '#%02x%02x%02x'%(c(),c(),c())
        self.item = self.canvas.create_oval(x1, y1, x2, y2, fill=self.color, outline=self.color1)
            
    def move_ball(self): 
        self.xpos += self.xmove
        self.ypos += self.ymove
        if (self.xpos - self.radius)<=0 or (self.xpos + self.radius)>=self.canvas_w:
            self.xmove *= -1
        elif (self.ypos - self.radius)<=0 or (self.ypos + self.radius)>=self.canvas_h:
            self.ymove *= -1            
        self.canvas.move(self.item, self.xmove, self.ymove)
        
        
class ScreenSaver(): 
    balls = list()
    def __init__(self): 
        self.win_dow = tkinter.Tk()
        self.win_dow.overrideredirect(1)
        w = self.win_dow.winfo_screenwidth()
        h = self.win_dow.winfo_screenheight()
        self.canvas = tkinter.Canvas(self.win_dow, width=w, height=h)
        self.canvas.pack()
        self.balls_num = random.randint(10,25)
        for i in range(self.balls_num):
            ball = rball(self.canvas,w,h)
            self.balls.append(ball)            
        self.run_saver()        
        self.win_dow.bind('<Button-1>', self.levent1)
        self.win_dow.bind('<Key>', self.levent2)        
        self.win_dow.mainloop()
        
    def run_saver(self):
        for ball in self.balls:
            ball.move_ball()
        self.canvas.after(200,self.run_saver)
        
    def levent1(self,event):
        self.win1 = tkinter.Tk()
        self.btn = tkinter.Button(self.win1, text='鼠标事件', font=26,command=self.ljbquit)
        self.btn.pack()
        self.win1.mainloop()

    def levent2(self,event):
        self.win1 = tkinter.Tk()
        self.btn = tkinter.Button(self.win1, text='键盘事件', font=26,command=self.ljbquit)
        self.btn.pack()        
        self.win1.mainloop()
        
    def ljbquit(self):
        self.win_dow.destroy()

if __name__ == '__main__':
    ScreenSaver()
```
